```python
import numpy as np
from matplotlib import pyplot as plt
#Função é x^2 +2x e suas raízes são 0 e -2
#Método de Newton


#Função
def f(x):
  return x**2 + 2 * x

#Derivada
def g(x):
  return 2*x + 2

vetor=[]
t = range(1,10)

#Chutes iniciais
x = 4
#x= -4

for n in range (1,10):
  x = x - f(x)/g(x)  # aqui altera-se para o método do ponto fixo também
  print(x)
  vetor.append(x)
plt.plot(vetor,t)
```
